# Webscraping Data

Taken from Google Play Store

```python
from google_play_scraper import Sort, reviews
```

```python
#Extract reviews
result1, continuation_token = reviews(
    'com.netflix.mediaclient',
    lang='en', # defaults to 'en'
    country='us', # defaults to 'us'
    sort=Sort.NEWEST, # defaults to Sort.NEWEST
    count=7000, # defaults to 100
     # defaults to None(means all score)
)

result2, continuation_token = reviews(
    'com.wbd.stream',
    lang='en', # defaults to 'en'
    country='us', # defaults to 'us'
    sort=Sort.NEWEST, # defaults to Sort.NEWEST
    count=7000, # defaults to 100
     # defaults to None(means all score)
)

result3, continuation_token = reviews(
    'com.disney.disneyplus',
    lang='en', # defaults to 'en'
    country='us', # defaults to 'us'
    sort=Sort.NEWEST, # defaults to Sort.NEWEST
    count=7000, # defaults to 100
     # defaults to None(means all score)
) 

result4, continuation_token = reviews(
    'com.amazon.avod.thirdpartyclient',
    lang='en', # defaults to 'en'
    country='us', # defaults to 'us'
    sort=Sort.NEWEST, # defaults to Sort.NEWEST
    count=7000, # defaults to 100
     # defaults to None(means all score)
) 
```

Convert to Dataframe

```python
import numpy as np
import pandas as pd
```

```python
netflix = pd.DataFrame.from_dict(result1)
netflix['App'] = 'netflix'
hbo = pd.DataFrame.from_dict(result2)
hbo['App'] = 'HBO'
disney = pd.DataFrame.from_dict(result3)
disney['App'] = 'Disney+'
amazon = pd.DataFrame.from_dict(result4)
amazon['App'] = 'Amazon'
```

# Merge into 1 table

```python
frames = [netflix, disney, hbo, amazon]
df1 = pd.concat(frames)
display(df1)
```

```python
df1['score'] = df1['score'].apply(pd.to_numeric)
```

```python
def sentiment(x):
    sent = None 
    if x > 3.5:
        sent = "positive"
    elif x < 2.5:
        sent = "negative"
    else:
        sent = "neutral"
    return sent
```

```python
df1["act_label"] = df1["score"].apply(sentiment)
```
